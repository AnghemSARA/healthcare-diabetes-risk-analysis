```python
# This code checks unclear and missing values before making cleaning decisions

import pandas as pd
import numpy as np

unclear_codes = {
    "DIABETE3": [2, 4, 7, 9],
    "GENHLTH": [7, 9],
    "PHYSHLTH": [77, 99],
    "MENTHLTH": [77, 99],
    "BLOODCHO": [7, 9],
    "CHOLCHK": [7, 9],
    "CVDINFR4": [7, 9],
    "CVDCRHD4": [7, 9],
    "CVDSTRK3": [7, 9],
    "DIFFWALK": [7, 9],
    "SMOKE100": [7, 9],
    "EXERANY2": [7, 9],
    "SEX": [7, 9],
    "EDUCA": [9],
    "INCOME2": [77, 99],
    "_AGEG5YR": [14]
}

summary = []

for col, codes in unclear_codes.items():
    unclear_count = df_selected[col].isin(codes).sum()
    missing_count = df_selected[col].isnull().sum()
    total_unclear_missing = unclear_count + missing_count
    percentage = (total_unclear_missing / len(df_selected)) * 100
    
    summary.append({
        "Column": col,
        "Unclear_values": unclear_count,
        "Missing_values": missing_count,
        "Total_unclear_or_missing": total_unclear_missing,
        "Percentage": round(percentage, 2)
    })

# Add BMI separately because BMI uses real missing values
bmi_missing = df_selected["_BMI5"].isnull().sum()

summary.append({
    "Column": "_BMI5",
    "Unclear_values": 0,
    "Missing_values": bmi_missing,
    "Total_unclear_or_missing": bmi_missing,
    "Percentage": round((bmi_missing / len(df_selected)) * 100, 2)
})

missing_summary_df = pd.DataFrame(summary)

missing_summary_df.sort_values(by="Percentage", ascending=False)
```

```python
# This code rebuilds df_clean 

df_clean = df_selected.copy()

# 1. Clean target column: keep only clear diabetes answers
# 1 = Diabetes, 3 = No diabetes
df_clean = df_clean[df_clean["DIABETE3"].isin([1, 3])]

df_clean["Diabetes_binary"] = df_clean["DIABETE3"].map({
    1: 1,
    3: 0
})

# 2. Keep unknown categories for high missing columns
# INCOME2: 77 and 99 become 9 = Unknown / Not reported
df_clean["INCOME2"] = df_clean["INCOME2"].replace({
    77: 9,
    99: 9
})

# CHOLCHK: 7 and 9 become 3 = Unknown / Not reported
df_clean["CHOLCHK"] = df_clean["CHOLCHK"].replace({
    7: 3,
    9: 3
})

# EXERANY2: 7 and 9 become 3 = Unknown / Not reported
df_clean["EXERANY2"] = df_clean["EXERANY2"].replace({
    7: 3,
    9: 3
})

# 3. Clean health columns
df_clean["GENHLTH"] = df_clean["GENHLTH"].replace({
    7: np.nan,
    9: np.nan
})

df_clean["PHYSHLTH"] = df_clean["PHYSHLTH"].replace({
    77: np.nan,
    88: 0,
    99: np.nan
})

df_clean["MENTHLTH"] = df_clean["MENTHLTH"].replace({
    77: np.nan,
    88: 0,
    99: np.nan
})

# 4. Clean low-missing binary columns
low_missing_binary_columns = [
    "BLOODCHO",
    "CVDINFR4",
    "CVDCRHD4",
    "CVDSTRK3",
    "DIFFWALK",
    "SMOKE100"
]

for col in low_missing_binary_columns:
    df_clean[col] = df_clean[col].replace({
        7: np.nan,
        9: np.nan
    })

# 5. Clean demographic columns
df_clean["EDUCA"] = df_clean["EDUCA"].replace({
    9: np.nan
})

df_clean["_AGEG5YR"] = df_clean["_AGEG5YR"].replace({
    14: np.nan
})

# 6. Remove rows with missing values only in low-missing columns
columns_to_drop_missing = [
    "GENHLTH",
    "PHYSHLTH",
    "MENTHLTH",
    "BLOODCHO",
    "CVDINFR4",
    "CVDCRHD4",
    "CVDSTRK3",
    "DIFFWALK",
    "SMOKE100",
    "SEX",
    "EDUCA",
    "_AGEG5YR"
]

print("Shape before final dropping:", df_clean.shape)

df_clean = df_clean.dropna(subset=columns_to_drop_missing)

print("Shape after final dropping:", df_clean.shape)
print("Remaining missing values:")
print(df_clean.isnull().sum())
```

```python
# This code replaces remaining missing values in CHOLCHK and EXERANY2 with Unknown category

df_clean["CHOLCHK"] = df_clean["CHOLCHK"].fillna(3)
df_clean["EXERANY2"] = df_clean["EXERANY2"].fillna(3)

# Check the result
print("Missing values in CHOLCHK:", df_clean["CHOLCHK"].isnull().sum())
print("Missing values in EXERANY2:", df_clean["EXERANY2"].isnull().sum())

print("\nCHOLCHK values:")
print(df_clean["CHOLCHK"].value_counts().sort_index())

print("\nEXERANY2 values:")
print(df_clean["EXERANY2"].value_counts().sort_index())
```

```python
# This code checks the remaining columns that still have missing values

remaining_columns = [
    "ALCDAY5",
    "FRUITJU1",
    "FRUIT1",
    "VEGETAB1",
    "_BMI5"
]

for col in remaining_columns:
    print("\nColumn:", col)
    print("Missing values:", df_clean[col].isnull().sum())
    print("Unique values sample:")
    print(sorted(df_clean[col].dropna().unique())[:40])
```

```python
# This code converts _BMI5 into a readable BMI value and fills missing BMI using the median

df_clean["BMI"] = df_clean["_BMI5"] / 100

bmi_median = df_clean["BMI"].median()

df_clean["BMI"] = df_clean["BMI"].fillna(bmi_median)

print("BMI median used:", bmi_median)
print("Missing BMI after cleaning:", df_clean["BMI"].isnull().sum())

df_clean[["_BMI5", "BMI"]].head()
```

```python
# This code shows the full value counts of alcohol consumption codes

df_clean["ALCDAY5"].value_counts(dropna=False).sort_index()
```

```python
# This code converts ALCDAY5 into estimated alcohol drinking days per month

def convert_alcohol_days(value):
    if pd.isna(value):
        return np.nan
    
    value = int(value)
    
    # 888 means no alcohol consumption
    if value == 888:
        return 0
    
    # 777 and 999 mean unknown/refused
    if value in [777, 999]:
        return np.nan
    
    # Codes 101-107 mean days per week
    # Convert weekly frequency to monthly frequency by multiplying by 4
    if 101 <= value <= 107:
        days_per_week = value - 100
        return days_per_week * 4
    
    # Codes 201-230 mean days per month
    if 201 <= value <= 230:
        days_per_month = value - 200
        return days_per_month
    
    return np.nan


df_clean["Alcohol_days_per_month"] = df_clean["ALCDAY5"].apply(convert_alcohol_days)

print("Missing values before imputation:", df_clean["Alcohol_days_per_month"].isnull().sum())

alcohol_median = df_clean["Alcohol_days_per_month"].median()

df_clean["Alcohol_days_per_month"] = df_clean["Alcohol_days_per_month"].fillna(alcohol_median)

print("Alcohol median used:", alcohol_median)
print("Missing values after imputation:", df_clean["Alcohol_days_per_month"].isnull().sum())

df_clean[["ALCDAY5", "Alcohol_days_per_month"]].head(10)
```

```python
# This code converts BRFSS fruit and vegetable frequency codes into times per day

def convert_food_frequency(value):
    if pd.isna(value):
        return np.nan
    
    value = int(value)
    
    # 777 = Don't know, 999 = Refused
    if value in [777, 999]:
        return np.nan
    
    # 300 means less than once per month
    if value == 300:
        return 0
    
    # Codes 101-199 mean times per day
    if 101 <= value <= 199:
        return value - 100
    
    # Codes 201-299 mean times per week
    # Convert weekly frequency to daily frequency
    if 201 <= value <= 299:
        return (value - 200) / 7
    
    # Codes 301-399 mean times per month
    # Convert monthly frequency to daily frequency
    if 301 <= value <= 399:
        return (value - 300) / 30
    
    return np.nan


df_clean["Fruit_juice_per_day"] = df_clean["FRUITJU1"].apply(convert_food_frequency)
df_clean["Fruit_per_day"] = df_clean["FRUIT1"].apply(convert_food_frequency)
df_clean["Vegetables_per_day"] = df_clean["VEGETAB1"].apply(convert_food_frequency)

food_columns = ["Fruit_juice_per_day", "Fruit_per_day", "Vegetables_per_day"]

print("Missing values before imputation:")
print(df_clean[food_columns].isnull().sum())

for col in food_columns:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_value)
    print(f"{col} median used:", median_value)

print("\nMissing values after imputation:")
print(df_clean[food_columns].isnull().sum())

df_clean[["FRUITJU1", "Fruit_juice_per_day", "FRUIT1", "Fruit_per_day", "VEGETAB1", "Vegetables_per_day"]].head(10)
```

```python
# This code correctly converts BRFSS fruit and vegetable frequency codes into times per day

def convert_food_frequency(value):
    if pd.isna(value):
        return np.nan
    
    value = int(value)
    
    # 555 = Never
    if value == 555:
        return 0
    
    # 777 = Don't know, 999 = Refused
    if value in [777, 999]:
        return np.nan
    
    # 300 = Less than once per month
    if value == 300:
        return 0
    
    # 101-199 = times per day
    if 101 <= value <= 199:
        return value - 100
    
    # 201-299 = times per week
    if 201 <= value <= 299:
        return (value - 200) / 7
    
    # 301-399 = times per month
    if 301 <= value <= 399:
        return (value - 300) / 30
    
    return np.nan


df_clean["Fruit_juice_per_day"] = df_clean["FRUITJU1"].apply(convert_food_frequency)
df_clean["Fruit_per_day"] = df_clean["FRUIT1"].apply(convert_food_frequency)
df_clean["Vegetables_per_day"] = df_clean["VEGETAB1"].apply(convert_food_frequency)

food_columns = ["Fruit_juice_per_day", "Fruit_per_day", "Vegetables_per_day"]

print("Missing values before imputation:")
print(df_clean[food_columns].isnull().sum())

for col in food_columns:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_value)
    print(f"{col} median used:", median_value)

print("\nMissing values after imputation:")
print(df_clean[food_columns].isnull().sum())

df_clean[["FRUITJU1", "Fruit_juice_per_day", 
          "FRUIT1", "Fruit_per_day", 
          "VEGETAB1", "Vegetables_per_day"]].head(10)
```
